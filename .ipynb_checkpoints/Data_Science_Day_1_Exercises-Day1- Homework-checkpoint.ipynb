```python
"""
Example Result - Not the actual result
Boston 18 had more SLG in 2018
[0.404 0.443 0.392 0.403 0.482 0.424 0.402 0.459 0.429 0.354 0.387 0.333
 0.539 0.243 0.262 0.354 0.342 0.339 0.328 0.306 0.222 0.2   1.   ]
"""

FIELDS = ['Rk', 'Pos', 'Name', 'Age', 'G', 'PA', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI', 'SB', 'CS', 
          'BB', 'SO', 'BA', 'OBP', 'SLG', 'OPS', 'OPS+', 'TB', 'GDP', 'HBP', 'SH', 'SF', 'IBB']

DATATYPES = [('rk', 'i'), ('pos', '|S25'), ('name', '|S25'), ('age', 'i'), ('g', 'i'), ('pa', 'i'), ('ab', 'i'),
                ('r', 'i'), ('h', 'i'), ('2b', 'i'), ('3b', 'i'), ('hr', 'i'), ('rbi', 'i'), ('sb', 'i'), ('cs', 'i'),
                ('bb', 'i'), ('so', 'i'), ('ba', 'f'), ('obp', 'f'), ('slg', 'f'), ('ops', 'f'), ('opsp', 'i'),
                ('tb', 'i'), ('gdp', 'i'), ('hbp', 'i'), ('sh', 'i'), ('sf', 'i'), ('ibb', 'i')]

#Use this function to actually open the data inside of a NDArray

def load_data(filename, d = ','):
    data = np.genfromtxt(filename, delimiter = d, skip_header = 1,
                        usecols = np.arange(0,24), invalid_raise = False,
                        names = FIELDS, dtype = DATATYPES)
    return data

bs2017 = load_data('../../files/redsox_2017_hitting.txt')
bs2018 = load_data('../../files/redsox_2018_hitting.txt')

all_slg2017 = bs2017['SLG']
all_names2017 = bs2017['Name']
all_slg2018 = bs2018['SLG']
all_names2018 = bs2018['Name']

data2017 = np.array(bs2017['SLG'].shape)
data2018 = np.array(bs2018['SLG'].shape)


# Construct 2017

names_2017 = []

for name in range(len(all_names2017)):
    names_2017.append(all_names2017[name].decode('utf-8'))
    
zipped_values2017 = list(zip(names_2017,all_slg2017))

# print(zipped_values2017)

sorted_2017 = sorted(zipped_values2017, key = lambda x: x[0], reverse = False)
print(sorted_2017)



print('\n')

# Construct 2018

names_2018 = []

for name in range(len(all_names2018)):
    names_2018.append(all_names2018[name].decode('utf-8'))
    
zipped_values2018 = list(zip(names_2018,all_slg2018))

# print('\n',zipped_values2018)

sorted_2018 = sorted(zipped_values2018, key = lambda x: x[0], reverse = False)
print(sorted_2018)




def findBetter():
    better_slg = []
    for i in sorted_2017:
        for j in sorted_2018:
            if i[1] > j[1]:
                better_slg.append('2017')
            else:
                better_slg.append('2018')
            
    return better_slg

print(better_slg)
            


```
